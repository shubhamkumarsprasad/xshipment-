```python
import json

def extract_datetime(row):
    milestones = row['milestones']
    datetime_list = []
    try:
        milestones_list = json.loads(milestones)
        for milestone in milestones_list:
            if milestone.get('value') in ['Booking Confirmed', 'Delivered to Consignee']:
                datetime_list.append(milestone.get('dateTime'))
    except (json.JSONDecodeError, TypeError):
        pass  # Handle JSON parsing error or missing values gracefully
    return datetime_list

# Check for missing values in the 'milestones' column
df = df.dropna(subset=['milestones'])

# Apply the function to each row
df['datetime'] = df.apply(extract_datetime, axis=1)

# Flatten the list of datetimes
df = df.explode('datetime')
```

```python
import json

def extract_datetime(row):
    milestones = row['milestones']
    booking_confirmed_datetime = None
    delivered_datetime = None
    try:
        milestones_list = json.loads(milestones)
        for milestone in milestones_list:
            if milestone.get('value') == 'Booking Confirmed':
                booking_confirmed_datetime = milestone.get('dateTime')
            elif milestone.get('value') == 'Delivered to Consignee':
                delivered_datetime = milestone.get('dateTime')
    except (json.JSONDecodeError, TypeError):
        pass  # Handle JSON parsing error or missing values gracefully
    return booking_confirmed_datetime, delivered_datetime

# Check for missing values in the 'milestones' column
df = df.dropna(subset=['milestones'])

# Apply the function to each row
df[['booking_confirmed_datetime', 'delivered_datetime']] = df.apply(extract_datetime, axis=1, result_type='expand')

```

```python
# Convert booking_confirmed_datetime and delivered_datetime to pandas datetime objects
df['booking_confirmed_datetime'] = pd.to_datetime(df['booking_confirmed_datetime'])
df['delivered_datetime'] = pd.to_datetime(df['delivered_datetime'])

# Extract year, month, and time components
df['booking_confirmed_year'] = df['booking_confirmed_datetime'].dt.year
df['booking_confirmed_month'] = df['booking_confirmed_datetime'].dt.month
df['booking_confirmed_time'] = df['booking_confirmed_datetime'].dt.time

df['delivered_year'] = df['delivered_datetime'].dt.year
df['delivered_month'] = df['delivered_datetime'].dt.month
df['delivered_time'] = df['delivered_datetime'].dt.time

# Drop the original datetime columns if needed
# df.drop(columns=['booking_confirmed_datetime', 'delivered_datetime'], inplace=True)
```

```python
import pandas as pd

# Group by 'kind' and 'delivered_month', then calculate the average delivery time
average_delivery_time_per_kind = df.groupby(['kind', 'delivered_month'])['delivery_days'].mean().reset_index()

# Rename the column for clarity
average_delivery_time_per_kind.rename(columns={'delivery_days': 'average_delivery_time_per_kind'}, inplace=True)

# Merge the calculated averages back into the original dataframe
df = df.merge(average_delivery_time_per_kind, on=['kind', 'delivered_month'], how='left')

```
